#### Handling Outliers

```python
mq1, mq2, mq3, mq4, mq5 = np.percentile(df["math score"], [0, 25, 50, 75, 100])
rq1, rq2, rq3, rq4, rq5 = np.percentile(df["reading score"], [0, 25, 50, 75, 100])
wq1, wq2, wq3, wq4, wq5 = np.percentile(df["writing score"], [0, 25, 50, 75, 100])
print("Math Score Quartiles:", mq1, mq2, mq3, mq4, mq5)
print("Reading Score Quartiles:", rq1, rq2, rq3, rq4, rq5)
print("Writing Score Quartiles:", wq1, wq2, wq3, wq4, wq5)
```

```python
m_iqr = mq4 - mq2
r_iqr = rq4 - rq2
w_iqr = wq4 - wq2
m_lower_bound = mq2 - 1.5 * m_iqr
m_upper_bound = mq4 + 1.5 * m_iqr
r_lower_bound = rq2 - 1.5 * r_iqr
r_upper_bound = rq4 + 1.5 * r_iqr
w_lower_bound = wq2 - 1.5 * w_iqr
w_upper_bound = wq4 + 1.5 * w_iqr
print("Math Score Outlier Bounds:", m_lower_bound, m_upper_bound)
print("Reading Score Outlier Bounds:", r_lower_bound, r_upper_bound)
print("Writing Score Outlier Bounds:", w_lower_bound, w_upper_bound)
```

```python
math_outliers = df[(df["math score"] < m_lower_bound) | (df["math score"] > m_upper_bound)]
reading_outliers = df[(df["reading score"] < r_lower_bound) | (df["reading score"] > r_upper_bound)]
writing_outliers = df[(df["writing score"] < w_lower_bound) | (df["writing score"] > w_upper_bound)]
print("Math Score Outliers:\n", math_outliers)
print("Reading Score Outliers:\n", reading_outliers)
print("Writing Score Outliers:\n", writing_outliers)
```

```python
df_encoded = pd.get_dummies(df, columns=[
    'gender',
    'race/ethnicity',
    'parental level of education',
    'lunch',
    'test preparation course'
])

df_encoded.head()
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['math score', 'reading score', 'writing score']

df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

df_encoded.head()
```

#### Data Transformation

```python
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

df['math_log'] = np.log1p(df['math score'])
df['reading_log'] = np.log1p(df['reading score'])
df['writing_log'] = np.log1p(df['writing score'])

df.head()
```

#### Importing the Modules and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("housing.csv")
```

#### Q1. Apply basic data exploration methods to understand the structure and summary of the dataset.


```python
dataset.head()
```

```python
dataset.info()
```

```python
prices = dataset["Price_USD"]
prices.head()
```

```python
prices.dtype
```

#### Q2. Calculate the **Mean** to determine the overall central value of the dataset.


```python
mean = prices.mean()
print(f"The mean price is: ${mean}")
```

#### Q3. Compute the **Median** to identify the middle value and analyze skewness.


```python
median = prices.median()
print(f"The median price is: ${median}")
```

#### Q4. Determine the **Mode** to find the most frequently occurring value (if applicable).


```python
mode = prices.mode()
print(f"The mode price is: ${mode[0]}")
```

#### Q5. Calculate the **Range** (difference between maximum and minimum values).


```python
max, min = prices.max(), prices.min()
range = max-min
print(f"The range for the prices is: {range}")
```

#### Q6. Compute the **Variance** to measure how data points deviate from the mean.


```python
variance = prices.var()
print(f"The Variance of the prices is: {variance}")
```

#### Q7. Calculate the **Standard Deviation** to understand the average spread of values.


```python
standard_deviation = prices.std()
print(f"The Standard Deviation of the prices is: {standard_deviation}")
```

#### Q8. Determine the **Interquartile Range (IQR)** to analyze the spread of the middle 50% of data.


```python
q1, q2, q3 = prices.quantile(.25), prices.quantile(.5), prices.quantile(.75)
iqr = q3-q1
print(f"The IQR (Inter-Quartile Range) of the prices is: {iqr}")
```

#### Q9. Analyze the **distribution shape** of the dataset (symmetric, skewed, presence of outliers, kurtosis).

```python
skewness = prices.skew()
print(f"Skewness: {skewness}")

if skewness < 0:
    print(f"The distribution is Left Skewed (Negatively Skew)")
elif skewness > 0:
    print(f"The distribution is Right Skewed (Positively Skewed)")
else:
    print(f"The distribution is Symmetric")
```

```python
kurtosis = prices.kurt()
print(f"Kurtosis: {kurtosis}")

if kurtosis < 0:
    print(f"This is Platykurtic Distribution")
elif kurtosis > 0:
    print(f"This is Leptokurtic Distribution")
else:
    print(f"This is Mesokurtic Distribution")
```

#### Q10. Support the analysis using visualizations such as:
- Histograms  
- Box plots  
- Scatter plots

```python
sns.histplot(prices, kde=True)
plt.title("Histrogram Distribution Plot")
plt.xlabel("Prices")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(prices, vert=False, showmeans=True)
plt.title("Boxplot of Prices")
plt.xlabel("Prices")
plt.text(mean, 1.1, f'Mean: ${mean:.2f}', color='green', ha='center', fontweight='bold')
plt.text(median, 0.85, f'Median: ${median:.2f}', color='orange', ha='center', fontweight='bold')
plt.show()
```

```python
sns.scatterplot(prices)
plt.title("Scatter Plot of Distribution")
plt.xlabel("Index")
plt.ylabel("Prices")
plt.show()
```

#### Q11. Compare Mean and Median to comment on skewness.

```python
print(f"Mean: ${mean} and Median: ${median}")

if mean < median:
    print(f"The Distribution is Left Skewed (Negatively Skewed)")
elif mean > median:
    print(f"The Distribution is Right Skewed (Positively Skewed)")
else:
    print(f"The Distribution is Symmetric")
```

#### Q12. Analyze variability using variance and standard deviation.

```python
print(f"Variance: {variance} and Standard Devistion: {standard_deviation}")

if standard_deviation > 0:
    print("The Distribution has variation in prices")
else:
    print(f"The Distribution does not have variation in prices")
```

#### Q13. Draw meaningful conclusions about the dataset’s behavior and patterns.

```python
corr_matrix = dataset.corr(numeric_only=True)
corr_matrix
```

```python
sns.heatmap(corr_matrix)
plt.title("Heatmap of the Dataset")
plt.show()
```

```python
price_corr = corr_matrix['Price_USD'].sort_values(ascending=False)
price_corr = price_corr[price_corr < 1]
best_feature = price_corr.idxmax()
print(f"The best feature is \"{best_feature}\" with Correlation Ratio of {price_corr[best_feature]:.4f}")
```

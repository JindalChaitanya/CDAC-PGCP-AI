#### Q1. Most expensive car for each car company make.

```python
data.groupby('Make')['MSRP'].max()
```

#### Q2.Most expensive convertible vehicle for each company.

```python
data[data['Vehicle Style'] == 'Convertible'].groupby('Make')['MSRP'].max()
```

#### Q3.List all the cars which are automatic.

```python
data[data['Transmission Type'] == 'AUTOMATIC']
```

#### Q4. Premium cars of each car company

```python
data[data['Engine Fuel Type'] == "premium unleaded (required)"].groupby(['Make', 'Model'])[['Make', 'Model']].size()
```

#### Q5. Average price of cars each car company make.

```python
data.groupby(['Make', 'Model'])['MSRP'].mean()
```

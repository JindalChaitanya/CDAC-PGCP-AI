#### Q1. Find the Pokémon names for which **Legendary = True**.

```python
legendary_pokemon = dataset[dataset['is_legendary'] == 1]['name']
print(legendary_pokemon.to_list())
```

#### Q2. Take a Pokémon name as input from the user and print its complete details.

```python
input_name = input("Enter a Pokemon's Name")
print(f"Enter a Pokemon's Name: {input_name}")

pokemon = dataset[dataset['name'] == input_name.capitalize()]

if not pokemon.empty:
    print(pokemon)
else:
    print(f"No pokemon found...")
```

#### Q3. Find the Pokémon with the **maximum Total**.  
- If a tie exists, prioritize **Legendary = True**  
- If still tied, prioritize **Generation = 3**

```python
maxTotal = dataset['base_total'].max()
filtered = dataset[dataset['base_total'] == maxTotal]

if len(filtered) > 1:
    filtered = filtered.sort_values(by='is_legendary', ascending=False)
    gen3 = filtered[filtered['generation'] == 3]
    if not gen3.empty:
        filtered = gen3

result = filtered.iloc[0]

print(f"Pokemon with maximum Total: \n{result[['name', 'base_total', 'is_legendary', 'generation']]}")
```

#### Q1. Create a pandas DataFrame from a dictionary with the following data.

| Name    | Age | City     |

| Charlie | 22  | Paris    |

```python
import pandas as pd

students_df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 27, 22],
    "City": ["New York", "London", "Paris"]
})

filepath = "students.csv"
students_df.to_csv(filepath, index=False)

print("Dummy students.csv created:")
print(students_df)
print(f"Saved at: {filepath}")
```

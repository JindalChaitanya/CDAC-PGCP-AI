#### Q1. Create a pandas DataFrame from a dictionary with the following data.

Load the dataset and check head

```python
sal = pd.read_csv("Salaries.csv")
print(sal.head())
```

Make numeric columns safe for calculation

```python
numeric_cols = ["BasePay", "OvertimePay", "OtherPay", "Benefits", "TotalPay", "TotalPayBenefits", "Year"]
for col in numeric_cols:
    sal[col] = pd.to_numeric(sal[col], errors="coerce")
```

Use .info() to find number of entries

```python
sal.info()
```

Find average BasePay

```python
print("Average BasePay:", sal["BasePay"].mean())
```

Find highest OvertimePay

```python
print("Highest OvertimePay:", sal["OvertimePay"].max())
```

Find job title of JOSEPH DRISCOLL

```python
print(sal[sal["EmployeeName"] == "JOSEPH DRISCOLL"]["JobTitle"])
```

Find total pay including benefits of JOSEPH DRISCOLL

```python
print(sal[sal["EmployeeName"] == "JOSEPH DRISCOLL"]["TotalPayBenefits"])
```

Find highest paid person including benefits

```python
highest_paid = sal.loc[sal["TotalPayBenefits"].idxmax()]
print(highest_paid)
```

Find lowest paid person including benefits

```python
lowest_paid = sal.loc[sal["TotalPayBenefits"].idxmin()]
print(lowest_paid)
```

Find average BasePay per year (2011–2014)

```python
print(sal.groupby("Year")["BasePay"].mean())
```

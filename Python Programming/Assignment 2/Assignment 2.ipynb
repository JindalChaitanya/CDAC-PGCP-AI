#### Q8: Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself.
Sample String : 'restart'

Expected Result : 'resta$t'



```python
string = input("Enter a string: ")
print(f"Entered String: {string}")

if len(string) > 0:
    first_char = string[0]
    
    rest_string = string[1:]
    
    result = rest_string.replace(first_char, '$')
    
    print(f"Result: {first_char + result}")
    
```

#### Q10: Write a Python program to add 'ing' at the end of a given string (length should be at least 3). If the given string is already ends with 'ing' then add 'ly' instead.

```python
string = input("Enter a string: ")
print(f"Entered String: {string}")

if len(string) < 3:
    print(f"Output: {string}")

elif string[-3:] == 'ing':
    print(f"Output: {string[:-3]}ly")

else:
    print(f"Output: {string}ing")
```

#### Q11: The marks obtained by a student in 5 different Subjects are input through a keyboard.
        The Student gets a division as per the following rules.
         1. Percentage above or equal to 60 – First Division
         2. Percentage between 50 and 59 – Second Division
         3. Percentage between 40 and 49 – Third Division
         4. Percentage less than 40 – Fail
         Write a python program to Display the result based on the above Criteria.

```python
sub1 = float(input("Enter marks for Subject 1 (out of 100): "))
sub2 = float(input("Enter marks for Subject 2 (out of 100): "))
sub3 = float(input("Enter marks for Subject 3 (out of 100): "))
sub4 = float(input("Enter marks for Subject 4 (out of 100): "))
sub5 = float(input("Enter marks for Subject 5 (out of 100): "))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total_marks / 500) * 100

if percentage >= 60:
    division = "First Division"    
elif percentage >= 50 and percentage < 60:
    division = "Second Division"
elif percentage >= 40 and percentage < 50:
    division = "Third Division"
else:
    division = "Fail"
    
print(f"Total marks: {total_marks}")
print(f"Percentage: {percentage:.2f}%")
print(f"Final Result: {division}")
```

#### Q12: Write a Python program to find the largest number among the three input numbers.

```python
num1 = float(input("Enter the 1st number: "))
num2 = float(input("Enter the 2nd number: "))
num3 = float(input("Enter the 3rd number: "))

print(f"The numbers are {num1}, {num2}, {num3}")

if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

print(f"The largest number is {largest}")
```

#### Q13: Write a Python program to check if the input year is a leap year or not.

```python
year = int(input("Enter a year: "))

print(f"Entered year: {year}")

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"\n{year} is a leap year!")
else:
    print(f"\n{year} is NOT a leap year.")
```

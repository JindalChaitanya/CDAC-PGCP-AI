#### Q1. Write a Python program to read the first n lines of a file.


```python
def read_first_n_lines(file_name, n):
    file = open(file_name, "r")
    lines = file.readlines()
    file.close()

    print(f"First {n} lines of the file {file_name}:")
    for i in range(min(n, len(lines))):
        print(lines[i], end="")

file_name = input("Enter the file name: ") + '.txt'
n = int(input("Enter the number of lines to read: "))

read_first_n_lines(file_name, n)
```

#### Q2. Write a Python program to append text to a file and display the text.


```python
def append_and_display(file_name, text):
    file = open(file_name, "a")
    file.write(text + "\n")
    file.close()

    file = open(file_name, "r")
    content = file.read()
    file.close()

    print(f"Content of the file {file_name}:")
    print(content)

file_name = input("Enter the file name: ") + '.txt'
text = input("Enter the text to append: ")

append_and_display(file_name, text)
```

#### Q3. Write a Python program to read a file line by line and store it into a list.


```python
def read_file_into_list(file_name):
    file = open(file_name, "r")
    lines = file.readlines()
    file.close()

    print(f"List of lines in file {file_name}:")
    print(lines)

file_name = input("Enter the file name: ") + '.txt'

read_file_into_list(file_name)
```

#### Q4. Write a program to print each line of a file in reverse order.


```python
def print_lines_in_reverse(file_name):
    file = open(file_name, "r")
    lines = file.readlines()
    file.close()

    print(f"Lines of file {file_name} in reverse order:")
    for line in reversed(lines):
        print(line, end="")

file_name = input("Enter the file name: ") + '.txt'

print_lines_in_reverse(file_name)
```

#### Q5. Write a Python program to write list contents to a file.


```python
def write_list_to_file(file_name, items):
    file = open(file_name, "w")
    for item in items:
        file.write(item + "\n")
    file.close()

    file = open(file_name, "r")
    content = file.read()
    file.close()

    print(f"List contents written to the file {file_name}:")
    print(content)

file_name = input("Enter the file name: ") + '.txt'
items = input("Enter list items separated by commas: ").split(",")

items = [item.strip() for item in items]

write_list_to_file(file_name, items)
```

#### Q6. Write a program to compute the number of characters, words, and lines in a file.


```python
def count_file_contents(file_name):
    file = open(file_name, "r")
    text = file.read()
    file.close()

    char_count = len(text)
    word_count = len(text.split())
    line_count = len(text.splitlines())

    print(f"Number of characters in {file_name} = {char_count}")
    print(f"Number of words in {file_name} = {word_count}")
    print(f"Number of lines in {file_name} = {line_count}")

file_name = input("Enter the file name: ") + '.txt'

count_file_contents(file_name)
```

#### Q7. Write a program to accept sentences from the user until the user enters "END". Save the data in a text file and display only those sentences that begin with an uppercase alphabet.


```python
def store_and_display_uppercase_sentences(file_name):
    file = open(file_name, "w")

    while True:
        sentence = input("Enter a sentence (type END to stop): ")
        if sentence == "END":
            break
        file.write(sentence + "\n")

    file.close()

    file = open(file_name, "r")
    sentences = file.readlines()
    file.close()

    print("Sentences beginning with an uppercase alphabet:")
    for sentence in sentences:
        stripped_sentence = sentence.strip()
        if stripped_sentence and stripped_sentence[0].isupper():
            print(stripped_sentence)

file_name = input("Enter the file name: ") + '.txt'

store_and_display_uppercase_sentences(file_name)
```

#### Q8. Write a program to store the following records in a binary file:
- Item No (integer)
- Item_Name (string)
- Qty (integer)
- Price (float)

The number of records should be entered by the user. Display the records with calculated amount (Price × Qty).

```python
import pickle

def store_and_display_records(file_name):
    records = []

    number_of_records = int(input("Enter the number of records: "))

    for i in range(number_of_records):
        print(f"Enter details for record {i + 1}:")
        item_no = int(input("Enter item number: "))
        item_name = input("Enter item name: ")
        qty = int(input("Enter quantity: "))
        price = float(input("Enter price: "))

        record = {
            "item_no": item_no,
            "item_name": item_name,
            "qty": qty,
            "price": price
        }
        records.append(record)

    file = open(file_name, "wb")
    pickle.dump(records, file)
    file.close()

    file = open(file_name, "rb")
    stored_records = pickle.load(file)
    file.close()

    print(f"Records stored in {file_name}:")
    for record in stored_records:
        amount = record["qty"] * record["price"]
        print(f'Item No: {record["item_no"]}, Item Name: {record["item_name"]}, Qty: {record["qty"]}, Price: {record["price"]}, Amount: {amount}')

file_name = input("Enter the binary file name: ")

store_and_display_records(file_name)
```

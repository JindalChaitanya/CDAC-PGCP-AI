#### Q1. Define a class representing a bank account with attributes for depositor name, account number, account type, and balance. Implement methods for deposit, withdrawal, and display.

```python
class BankAccount:
    def __init__(self, depositor_name, account_number, account_type, balance):
        self.depositor_name = depositor_name
        self.account_number = account_number
        self.account_type = account_type
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited Amount: {amount}")
        print(f"Updated Balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient balance")
        else:
            self.balance -= amount
            print(f"Withdrawn Amount: {amount}")
            print(f"Updated Balance: {self.balance}")

    def display(self):
        print(f"\n--- Account Details ---")
        print(f"Depositor Name: {self.depositor_name}")
        print(f"Account Number: {self.account_number}")
        print(f"Account Type: {self.account_type}")
        print(f"Balance: {self.balance}")


name = input("Enter depositor name: ")
print(f"Enter depositor name: {name}")

acc_no = input("Enter account number: ")
print(f"Enter account number: {acc_no}")

acc_type = input("Enter account type: ")
print(f"Enter account type: {acc_type}")

balance = float(input("Enter initial balance: "))
print(f"Enter initial balance: {balance}")

account = BankAccount(name, acc_no, acc_type, balance)

while True:
    print(f"\n--- Bank Account Menu ---")
    print(f"1. Deposit")
    print(f"2. Withdraw")
    print(f"3. Display Account Details")
    print(f"4. Exit")

    choice = input("Enter your choice: ")
    print(f"Enter your choice: {choice}")

    if choice == "1":
        amount = float(input("Enter amount to deposit: "))
        print(f"Enter amount to deposit: {amount}")
        account.deposit(amount)

    elif choice == "2":
        amount = float(input("Enter amount to withdraw: "))
        print(f"Enter amount to withdraw: {amount}")
        account.withdraw(amount)

    elif choice == "3":
        account.display()

    elif choice == "4":
        print(f"Exiting program.")
        break

    else:
        print(f"Invalid choice. Please try again.")
```

#### Q2. Create a base class `Staff` with attributes code and name. Derive classes:
- Teacher (subject, publication)
- Typist (speed)
- Officer (grade)
#### Further derive classes Regular and Casual from Typist.

```python
class Staff:
    def __init__(self, code, name):
        self.code = code
        self.name = name

    def display_staff(self):
        print(f"Code: {self.code}")
        print(f"Name: {self.name}")


class Teacher(Staff):
    def __init__(self, code, name, subject, publication):
        super().__init__(code, name)
        self.subject = subject
        self.publication = publication

    def display_teacher(self):
        print(f"\n--- Teacher Details ---")
        self.display_staff()
        print(f"Subject: {self.subject}")
        print(f"Publication: {self.publication}")


class Typist(Staff):
    def __init__(self, code, name, speed):
        super().__init__(code, name)
        self.speed = speed

    def display_typist(self):
        print(f"\n--- Typist Details ---")
        self.display_staff()
        print(f"Speed: {self.speed} wpm")


class Regular(Typist):
    def __init__(self, code, name, speed, salary):
        super().__init__(code, name, speed)
        self.salary = salary

    def display_regular(self):
        print(f"\n--- Regular Typist Details ---")
        self.display_typist()
        print(f"Salary: {self.salary}")


class Casual(Typist):
    def __init__(self, code, name, speed, daily_wages):
        super().__init__(code, name, speed)
        self.daily_wages = daily_wages

    def display_casual(self):
        print(f"\n--- Casual Typist Details ---")
        self.display_typist()
        print(f"Daily Wages: {self.daily_wages}")


class Officer(Staff):
    def __init__(self, code, name, grade):
        super().__init__(code, name)
        self.grade = grade

    def display_officer(self):
        print(f"\n--- Officer Details ---")
        self.display_staff()
        print(f"Grade: {self.grade}")


teacher = Teacher("T101", "Anita Sharma", "Mathematics", "5 Publications")
typist_regular = Regular("R201", "Rahul Verma", 60, 30000)
typist_casual = Casual("C301", "Neha Singh", 55, 800)
officer = Officer("O401", "Amit Kumar", "A")

teacher.display_teacher()
typist_regular.display_regular()
typist_casual.display_casual()
officer.display_officer()
```

#### Q3. Create an abstract class `Appliance` with abstract methods `turn_on()` and `turn_off()`. Implement subclasses `Fan` and `Light`.

```python
from abc import ABC, abstractmethod

class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass


class Fan(Appliance):
    def turn_on(self):
        print(f"Fan is turned on.")

    def turn_off(self):
        print(f"Fan is turned off.")


class Light(Appliance):
    def turn_on(self):
        print(f"Light is turned on.")

    def turn_off(self):
        print(f"Light is turned off.")


fan = Fan()
light = Light()

fan.turn_on()
fan.turn_off()

light.turn_on()
light.turn_off()
```

#### Q4. Create an abstract class `Shape` with an abstract method `area()` and a regular method `display_shape()`. Implement it with `Circle` and `Rectangle`.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    def display_shape(self):
        print(f"This is a shape.")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def display_shape(self):
        print(f"This is a Circle.")
        print(f"Radius: {self.radius}")
        print(f"Area: {self.area():.2f}")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def display_shape(self):
        print(f"This is a Rectangle.")
        print(f"Length: {self.length}")
        print(f"Width: {self.width}")
        print(f"Area: {self.area():.2f}")


radius = float(input("Enter radius of circle: "))
print(f"Enter radius of circle: {radius}")

length = float(input("Enter length of rectangle: "))
print(f"Enter length of rectangle: {length}")

width = float(input("Enter width of rectangle: "))
print(f"Enter width of rectangle: {width}")

circle = Circle(radius)
rectangle = Rectangle(length, width)

print(f"\n--- Circle ---")
circle.display_shape()

print(f"\n--- Rectangle ---")
rectangle.display_shape()
```

```python
import sqlite3

conn = sqlite3.connect("sales.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE CLIENT_MASTER (
    CLIENTNO TEXT PRIMARY KEY CHECK (CLIENTNO LIKE 'C%'),
    NAME TEXT NOT NULL,
    CITY TEXT,
    PINCODE INTEGER,
    STATE TEXT,
    BALDUE REAL
)
""")
```

```python
cursor.execute("""
CREATE TABLE PRODUCT_MASTER (
    PRODUCTNO TEXT PRIMARY KEY CHECK (PRODUCTNO LIKE 'P%'),
    DESCRIPTION TEXT NOT NULL,
    PROFITPERCENT REAL NOT NULL,
    UNITMEASURE TEXT NOT NULL,
    QTYONHAND INTEGER NOT NULL,
    REORDERLVL INTEGER NOT NULL,
    SELLPRICE REAL NOT NULL CHECK (SELLPRICE > 0),
    COSTPRICE REAL NOT NULL CHECK (COSTPRICE > 0)
)
""")
```

```python
cursor.execute("""
CREATE TABLE SALESMAN_MASTER (
    SALESMANNO TEXT PRIMARY KEY CHECK (SALESMANNO LIKE 'S%'),
    SALESMANNAME TEXT NOT NULL,
    ADDRESS1 TEXT NOT NULL,
    ADDRESS2 TEXT,
    CITY TEXT,
    PINCODE INTEGER,
    STATE TEXT,
    SALAMT REAL NOT NULL CHECK (SALAMT > 0),
    TGTTOGET REAL NOT NULL,
    YTDSALES REAL NOT NULL,
    REMARKS TEXT
)
""")
```

```python
cursor.execute("""
CREATE TABLE SALES_ORDER (
    ORDERNO TEXT PRIMARY KEY CHECK (ORDERNO LIKE 'O%'),
    CLIENTNO TEXT,
    ORDERDATE TEXT NOT NULL,
    DELYADDR TEXT,
    SALESMANNO TEXT,
    DELYTYPE TEXT,
    BILLYN TEXT,
    DELYDATE TEXT,
    ORDERSTATUS TEXT,
    FOREIGN KEY (CLIENTNO) REFERENCES CLIENT_MASTER(CLIENTNO),
    FOREIGN KEY (SALESMANNO) REFERENCES SALESMAN_MASTER(SALESMANNO)
)
""")
```

```python
cursor.execute("""
CREATE TABLE SALES_ORDER_DETAILS (
    ORDERNO TEXT,
    PRODUCTNO TEXT,
    QTYORDERED INTEGER,
    QTYDISP INTEGER,
    PRODUCTRATE REAL,
    FOREIGN KEY (ORDERNO) REFERENCES SALES_ORDER(ORDERNO),
    FOREIGN KEY (PRODUCTNO) REFERENCES PRODUCT_MASTER(PRODUCTNO)
)
""")

conn.commit()
```

#### Q3. Generate SQL statements to perform the following computations.

a. List the names of all clients having **'a' as the second letter** in their names.

b. List the clients who stay in a **city whose first letter is 'M'**.

c. List all clients who stay in **Bangalore or Mangalore**.

d. List all clients whose **BalDue is greater than 10000**.

e. List all information from the **Sales_order table for orders placed in the month of June**.

f. List the **Order No and the day on which clients placed their order**.

g. List the **name, city and state of clients who are not in the state of Maharashtra**.

a. Second letter is 'a'

```python
cursor.execute("""
SELECT NAME FROM CLIENT_MASTER
WHERE NAME LIKE '_a%'
""")
print(cursor.fetchall())
```

b. City starts with 'M'

```python
cursor.execute("""
SELECT * FROM CLIENT_MASTER
WHERE CITY LIKE 'M%'
""")
print(cursor.fetchall())
```

c. Bangalore or Mangalore

```python
cursor.execute("""
SELECT * FROM CLIENT_MASTER
WHERE CITY IN ('Bangalore','Mangalore')
""")
print(cursor.fetchall())
```

d. BalDue > 10000

```python
cursor.execute("""
SELECT * FROM CLIENT_MASTER
WHERE BALDUE > 10000
""")
print(cursor.fetchall())
```

e. Orders in June

```python
cursor.execute("""
SELECT * FROM SALES_ORDER
WHERE strftime('%m', ORDERDATE) = '06'
""")
print(cursor.fetchall())
```

f. Order No and day

```python
cursor.execute("""
SELECT ORDERNO, strftime('%d', ORDERDATE) AS ORDER_DAY
FROM SALES_ORDER
""")
print(cursor.fetchall())
```

g. Not in Maharashtra

```python
cursor.execute("""
SELECT NAME, CITY, STATE FROM CLIENT_MASTER
WHERE STATE != 'Maharashtra'
""")
print(cursor.fetchall())
```

#### Q4. Exercises using HAVING, GROUP BY and JOINS.

a. Print the **description and total quantity sold for each product**.

b. Calculate the **average quantity sold for each client that has a maximum order value of 15000**.

a. Description and total quantity sold

```python
cursor.execute("""
SELECT P.DESCRIPTION, SUM(SOD.QTYORDERED) AS TOTAL_QTY_SOLD
FROM PRODUCT_MASTER P
JOIN SALES_ORDER_DETAILS SOD
ON P.PRODUCTNO = SOD.PRODUCTNO
GROUP BY P.DESCRIPTION
""")
print(cursor.fetchall())
```

b. Average quantity sold per client (max order value 15000)

```python
cursor.execute("""
SELECT C.NAME, AVG(SOD.QTYORDERED) AS AVG_QTY_SOLD
FROM CLIENT_MASTER C
JOIN SALES_ORDER SO
ON C.CLIENTNO = SO.CLIENTNO
JOIN SALES_ORDER_DETAILS SOD
ON SO.ORDERNO = SOD.ORDERNO
GROUP BY C.NAME
HAVING SUM(SOD.QTYORDERED * SOD.PRODUCTRATE) = 15000
""")
print(cursor.fetchall())
```

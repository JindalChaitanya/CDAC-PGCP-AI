#### Q1. Write a program to make a **Menu Driven Simple Calculator** that can add, subtract, multiply, and divide using functions.

```python
def add(x, y):
    return f"Result of {x} + {y} = {x + y}"

def sub(x, y):
    return f"Result of {x} - {y} = {x - y}"

def mult(x, y):
    return f"Result of {x} * {y} = {x * y}"

def div (x, y):
    return f"Result of {x} / {y} = {x / y}"

menu = "Select Operation:\n1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit"

while True:
    print(menu)
    choice = int(input("Enter choice (1/2/3/4/5): "))
    print(f"Entered input (1/2/3/4/5): {choice}")
    if choice == 5:
        False
        print("Exiting the code...")
        break
    elif choice <= 0 or choice >= 6:
        print("Enter a valid input...")
    else:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))

        if choice == 1:
            print(add(num1, num2))
        elif choice == 2:
            print(sub(num1, num2))
        elif choice == 3:
            print(mult(num1, num2))
        elif choice == 4:
            print(div(num1, num2))
```

#### Q3. Write a Python program to display the **Fibonacci sequence up to the n-th term using recursive functions**.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

terms  = int(input("Enter how many terms of fibonacci series you want? -> "))
print(f"Fibonacci series upto {terms} terms:-")
for i in range(terms):
    print(fibonacci(i), end=" ")
```

#### Q4. Write a Python program to **find the sum of natural numbers up to n using a recursive function**.

```python
def sumof(n):
    if n <= 1:
        return n
    else:
        sum = n + sumof(n-1)
        return sum
    
terms  = int(input("Enter how many terms sum do you want? -> "))
print(f"Sum of {terms} terms: {sumof(terms)}")
```

#### Q5. Write a version of a palindrome recognizer that also accepts phrase palindromes such as:
 - "Go hang a salami I'm a lasagna hog."
 - "Was it a rat I saw?"
 - "Step on no pets"
 - "Sit on a potato pan, Otis"
 - "Lisa Bonet ate no basil"
 - "Satan, oscillate my metallic sonatas"
 - "I roamed under it as a tired nude Maori"
 - "Rise to vote sir"
 - "Dammit, I'm mad!"

```python
def isPalindrome(phrase):
    cleaned_chars = [char.lower() for char in phrase if char.isalnum()]
    return cleaned_chars == cleaned_chars[::-1]

test_phrases = [
    "Go hang a salami I'm a lasagna hog.",
    "Was it a rat I saw?",
    "Step on no pets",
    "Sit on a potato pan, Otis",
    "Lisa Bonet ate no basil",
    "Satan, oscillate my metallic sonatas",
    "I roamed under it as a tired nude Maori",
    "Rise to vote sir",
    "Dammit, I'm mad!",
    "Not a palindrome!"
]

for phrase in test_phrases:
    result = isPalindrome(phrase)
    print(f"[{'PASS' if result else 'FAIL'}]: {phrase}")
```

#### Q6. A **pangram** is a sentence that contains all the letters of the English alphabet at least once.

    Example:  
    "The quick brown fox jumps over the lazy dog."

#### Write a function to **check whether a sentence is a pangram or not**.


```python
import string

def isPangram(phrase):
    alphabets = set(string.ascii_lowercase)
    sentence_chars = set(phrase.lower())
    return alphabets.issubset(sentence_chars)

test_sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Pack my box with five dozen liquor jugs.",
    "How quickly daft jumping zebras vex.",
    "Hello world, I am not a pangram.",
    "abcdefghijklmnopqrstuvwxyz"
]

for sentence in test_sentences:
    result = isPangram(sentence)
    print(f"[{'PASS' if result else 'FAIL'}]: {sentence}")
```

#### Q7 Define a function `overlapping()` that takes **two lists** and returns:
- `True` if they have **at least one member in common**
- `False` otherwise.

```python
def isOverlapping(list1, list2):
    return bool(set(list1) & set(list2))

list1 = [int(item) for item in input("Enter List 1 comma seperated: ").split(',')]
list2 = [int(item) for item in input("Enter List 2 comma seperated: ").split(',')]

print(f"Entered List 1: {list1}\nEntered List 2: {list2}")
print(f"Do they have Overlapping elements: {'Yes' if isOverlapping(list1, list2) else 'No'}")
```

#### Q8. Write a function `find_longest_word()` that takes a **list of words** and returns the **length of the longest word**.

```python
import re

def find_longest_word():
    userInput = input("Enter the list of words or sentence(s): ").strip()
    print(f"Given input: {userInput}")

    wordList = re.split(r'[,\s]+', userInput)
    print(f"Word List: {wordList}")

    longestWord = max(wordList, key=len)
    longestLength = len(longestWord)

    return longestWord, longestLength

word, length = find_longest_word()
print(f"The Longest word is '{word}' and it's length is: {length}")
```

#### Q9. Write a function `filter_long_words()` that takes:
- a **list of words**
- an **integer n**
#### and returns the **list of words longer than n**.

```python
def filter_long_words(wordsList, n):
    return [word for word in wordsList if len(word) > n]

wordsList = input("Enter list of words comma seperated: ").split(',')
maxLength = int(input("Enter the atleast length of the words: "))

print(f"Original List: {wordsList}")
print(f"Words longer than {maxLength} letters -> {filter_long_words(wordsList, maxLength)}")
```

#### Q12. Write a program to **count the number of times a character appears in a given string**.


```python
from collections import Counter

def countCharacters():
    text = input("Enter a string to count its characters: ").strip()
    print(f"Entered String: {text}")

    cleanedList = [char.lower() for char in text if char]
    charCount = dict(Counter(cleanedList))

    print(charCount)

countCharacters()
```

#### Q13. Write a function to **convert a number into words**.
    Example:
    Input - `876`
    Output - `Eight Seven Six`

```python
def num2words():

    mapping = {
        '0': 'Zero',
        '1': 'One',
        '2': 'Two',
        '3': 'Three',
        '4': 'Four',
        '5': 'Five',
        '6': 'Six',
        '7': 'Seven',
        '8': 'Eight',
        '9': 'Nine'
    }

    userInput = input("Enter a number to convert: ").strip()

    translatedWords = []

    for char in userInput:
        if char in mapping:
            translatedWords.append(mapping[char])
        elif char == '-':
            translatedWords.append("Minus")
        elif char == ".":
            translatedWords.append("Point")
        else:
            print(f"{char} is not a valid number...")
            return
    
    finalOutput = " ".join(translatedWords)
    print(f"Input: {userInput}")
    print(f"Output: {finalOutput}")

num2words()
```

#### Q15. Write a program to input your **friends’ names and phone numbers** and store them in a dictionary.

#### Perform the following operations:
a) Display all **names and phone numbers**  
b) Add a **new key-value pair**  
c) Delete a **particular friend**  
d) Modify the **phone number of an existing friend**  
e) Check if a **friend exists in the dictionary**  
f) Display the dictionary in **sorted order of names**

```python
def phoneBook():
    phonebook = {}
    
    while True:
        print("1. Add a new friend\n2. View all friends\n3. Delete a friend\n4. Modify a friend's number\n5. Check if a friend exists\n6. View phonebook in sorted order\n7. Exit")

        try:
            choice = int(input("Enter your choice (1-7): "))
        except ValueError:
            print("Invalid input. Please enter a number from 1 to 7.")
            continue

        if choice == 1:
            name = input("Enter friend's name: ").strip()
            if not name:
                print("Name cannot be blank, Please try again...")
                continue
                
            number = input(f"Enter phone number for {name}: ").strip()
            phonebook[name] = number
            print(f"{name} successfully added to the phonebook...")

        elif choice == 2:
            if not phonebook:
                print("The phonebook is empty, Please add data first...")
            else:
                print(f"{'Name':<15}\n{'Phone Number':>15}")
                for name, number in phonebook.items():
                    print(f"{name:<15} | {number:>15}")

        elif choice == 3:
            name = input("Enter the name of the friend to delete: ").strip()
            if name in phonebook:
                del phonebook[name]
                print(f"{name} has been succesfully deleted...")
            else:
                print(f"Error: {name} is not in the phonebook.")

        elif choice == 4:
            name = input("Enter the name of the friend to modify: ").strip()
            if name in phonebook:
                new_number = input(f"Enter the new number for {name}: ").strip()
                phonebook[name] = new_number
                print(f"Success! {name}'s number updated.")
            else:
                print(f"Error: {name} is not in the phonebook.")

        elif choice == 5:
            name = input("Enter the name to search for: ").strip()
            if name in phonebook:
                print(f"Result: {name} is in the phonebook! Number: {phonebook[name]}")
            else:
                print(f"Result: {name} is not in the phonebook.")

        elif choice == 6:
            if not phonebook:
                print("The phonebook is empty.")
            else:
                print(f"{'Name':<15} | {'Phone Number':>15}")
                for name in sorted(phonebook):
                    print(f"{name:<15} | {phonebook[name]:>15}")

        elif choice == 7:
            print("Exiting the code...")
            break
            
        else:
            print("Invalid Choice...")

phoneBook()
```

#### Q16. Write a program to take **roll number, name, and percentage of marks for n students**.
#### Create user-defined functions to:
- Accept details of **n students**
- **Search** details using roll number
- **Display** results of all students
- Find the **overall topper**
- Find the **subject toppers**
    
    Hint: Use a **dictionary where key = roll number and value = (name, percentage)**.

```python
SUBJECTS = ['Math', 'Science', 'English']

def accept_details():
    students = {}
    try:
        n = int(input("How many students do you want to enter? "))
    except ValueError:
        print("Invalid number...")
        return students

    for i in range(n):
        try:
            roll = int(input("Enter Roll Number: "))
            name = input("Enter Name: ").strip()
            
            marks = {}
            total_marks = 0
            
            for sub in SUBJECTS:
                m = float(input(f"Enter marks for {sub} (out of 100): "))
                marks[sub] = m
                total_marks += m
                
            percentage = total_marks / len(SUBJECTS)
            
            students[roll] = (name, percentage, marks)
            print(f"{name} added successfully...")
            
        except ValueError:
            print("Invalid input...")
            
    return students

def search_student(students):
    if not students:
        print("No data available. Please add students first.")
        return
        
    try:
        roll = int(input("Enter the Roll Number to search: "))
        if roll in students:
            name, percentage, marks = students[roll]
            print(f"Name: {name}")
            print(f"Percentage: {percentage:.2f}%")
            print(f"Marks: {marks}")
        else:
            print(f"Error: Roll Number {roll} not found.")
    except ValueError:
        print("Invalid Roll Number format...")

def display_all(students):
    if not students:
        print("No data available. Please add students first.")
        return
        
    print(f"{'Roll No':<10} | {'Name':<15} | {'Percentage':<12} | {'Marks'}")
    print("="*55)
    
    for roll, data in students.items():
        name, percentage, marks = data
        print(f"{roll:<10} | {name:<15} | {percentage:>9.2f}%   | {marks}")

def find_overall_topper(students):
    if not students:
        print("\nNo data available.")
        return
        
    top_student = max(students.items(), key=lambda item: item[1][1])
    
    roll = top_student[0]
    name = top_student[1][0]
    percentage = top_student[1][1]
    
    print(f"OVERALL TOPPER: {name} (Roll No: {roll}) with {percentage:.2f}%")

def find_subject_toppers(students):
    if not students:
        print("No data available.")
        return
        
    for sub in SUBJECTS:
        highest_marks = -1
        topper_name = ""
        
        for roll, data in students.items():
            name, percentage, marks = data
            if marks[sub] > highest_marks:
                highest_marks = marks[sub]
                topper_name = name
                
        print(f"{sub:<10}: {topper_name} ({highest_marks} marks)")

def main():
    school_db = {}
    
    while True:
        print("1. Add Students")
        print("2. Search Student by Roll No")
        print("3. Display All Students")
        print("4. Find Overall Topper")
        print("5. Find Subject Toppers")
        print("6. Exit")
        
        choice = input("Enter your choice (1-6): ").strip()
        
        if choice == '1':
            new_students = accept_details()
            school_db.update(new_students)
        elif choice == '2':
            search_student(school_db)
        elif choice == '3':
            display_all(school_db)
        elif choice == '4':
            find_overall_topper(school_db)
        elif choice == '5':
            find_subject_toppers(school_db)
        elif choice == '6':
            print("Exiting the code...")
            break
        else:
            print("Invalid choice. Please select from 1 to 6.")

main()
```

#### Q17. A **bank** performs operations like deposit and withdrawal.
#### Create a **menu driven program** with the following options:
- Open a **Savings Bank Account**
- **Deposit** money
- **Withdraw** money (consider minimum balance requirement)
- Enter **amount and period for Fixed Deposit**
- Display the **maturity amount for the customer**

#### Also analyze:
- Information required for opening a **savings account**
- **Interest rate** offered by the bank
- **Minimum balance requirement**
- **FD interest rates** (may differ for senior citizens)

```python
def open_account(acc):
    if acc["opened"]:
        print("An account is already active.")
        return
    
    acc["name"] = input("Enter customer name: ")
    try:
        initial_deposit = float(input(f"Enter initial deposit (Min {acc['min_balance']}): "))
        if initial_deposit >= acc["min_balance"]:
            acc["balance"] = initial_deposit
            acc["opened"] = True
            print(f"Account opened! Current Balance: {acc['balance']:.2f}")
        else:
            print(f"Minimum deposit must be {acc['min_balance']}.")
    except ValueError:
        print("Invalid input.")

def deposit(acc):
    if not acc["opened"]:
        print("Please open an account first.")
        return
        
    try:
        amount = float(input("Enter amount to deposit: "))
        if amount > 0:
            acc["balance"] += amount
            print(f"Deposited {amount:.2f}. New Balance: {acc['balance']:.2f}")
        else:
            print("Deposit amount must be greater than zero.")
    except ValueError:
        print("Invalid input.")

def withdraw(acc):
    if not acc["opened"]:
        print("Please open an account first.")
        return
        
    try:
        amount = float(input("Enter amount to withdraw: "))
        if amount > 0:
            if (acc["balance"] - amount) >= acc["min_balance"]:
                acc["balance"] -= amount
                print(f"Withdrawn {amount:.2f}. New Balance: {acc['balance']:.2f}")
            else:
                print(f"Must maintain a minimum balance of {acc['min_balance']:.2f}.")
        else:
            print("Withdrawal amount must be greater than zero.")
    except ValueError:
        print("Invalid input.")

def calculate_fd():
    try:
        principal = float(input("Enter FD amount: "))
        years = float(input("Enter investment period (years): "))
        is_senior = input("Are you a Senior Citizen (60+)? (y/n): ").strip().lower()
        
        rate = 7.0 if is_senior == 'y' else 6.5
        maturity = principal * ((1 + rate / 100) ** years)
        
        print(f"Principal: {principal:.2f} | Tenure: {years} yrs | Rate: {rate}% | Maturity: {maturity:.2f}")
    except ValueError:
        print("Invalid input.")

account_details = {
    "opened": False, 
    "name": "", 
    "balance": 0.0, 
    "min_balance": 1000.0
}

while True:
    print("Menu:\n1.Open A/c\n2.Deposit\n3.Withdraw\n4.FD Calc\n5.Exit")
    choice = input("Enter choice: ")
    
    if choice == '1':
        open_account(account_details)
    elif choice == '2':
        deposit(account_details)
    elif choice == '3':
        withdraw(account_details)
    elif choice == '4':
        calculate_fd()
    elif choice == '5':
        print("Exiting the code...")
        break
    else:
        print("Invalid choice...")
```

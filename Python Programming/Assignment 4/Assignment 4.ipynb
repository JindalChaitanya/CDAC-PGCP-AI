#### Q1. Write a program to make a **Menu Driven Simple Calculator** that can add, subtract, multiply, and divide using functions.

```python
def add(x, y):
    return f"Result of {x} + {y} = {x + y}"

def sub(x, y):
    return f"Result of {x} - {y} = {x - y}"

def mult(x, y):
    return f"Result of {x} * {y} = {x * y}"

def div (x, y):
    return f"Result of {x} / {y} = {x / y}"

menu = "Select Operation:\n1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit"

while True:
    print(menu)
    choice = int(input("Enter choice (1/2/3/4/5): "))
    print(f"Entered input (1/2/3/4/5): {choice}")
    if choice == 5:
        False
        print("Exiting the code...")
        break
    elif choice <= 0 or choice >= 6:
        print("Enter a valid input...")
    else:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))

        if choice == 1:
            print(add(num1, num2))
        elif choice == 2:
            print(sub(num1, num2))
        elif choice == 3:
            print(mult(num1, num2))
        elif choice == 4:
            print(div(num1, num2))
```

#### Q3. Write a Python program to display the **Fibonacci sequence up to the n-th term using recursive functions**.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

terms  = int(input("Enter how many terms of fibonacci series you want? -> "))
print(f"Fibonacci series upto {terms} terms:-")
for i in range(terms):
    print(fibonacci(i), end=" ")
```

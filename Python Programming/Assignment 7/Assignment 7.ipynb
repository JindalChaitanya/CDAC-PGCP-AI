#### Q1. Subtract a week (7 days) from a given date in Python.

```python
from datetime import datetime, timedelta

date_str = input("Enter a date (YYYY-MM-DD): ")
print(f"Enter a date (YYYY-MM-DD): {date_str}")

given_date = datetime.strptime(date_str, "%Y-%m-%d")
result = given_date - timedelta(days=7)

print(f"Original date: {given_date.strftime('%Y-%m-%d')}")
print(f"Date after subtracting 7 days: {result.strftime('%Y-%m-%d')}")
```

#### Q2. Add one week (7 days) and 12 hours to a given date.


```python
from datetime import datetime, timedelta

date_str = input("Enter a date and time (YYYY-MM-DD HH:MM:SS): ")
print(f"Enter a date and time (YYYY-MM-DD HH:MM:SS): {date_str}")

given_date = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
result = given_date + timedelta(days=7, hours=12)

print(f"Original date and time: {given_date}")
print(f"Date after adding 7 days and 12 hours: {result}")
```

#### Q3. Print ten dates, each two weeks apart, starting from today in the format YYYY-MM-DD.

```python
from datetime import datetime, timedelta

start_date = datetime.today()
print(f"Starting date: {start_date.strftime('%Y-%m-%d')}")

for i in range(10):
    new_date = start_date + timedelta(weeks=2 * i)
    print(f"Date {i + 1}: {new_date.strftime('%Y-%m-%d')}")
```

#### Q4. Calculate the number of days between two given dates.

```python
from datetime import datetime

date1_str = input("Enter first date (YYYY-MM-DD): ")
print(f"Enter first date (YYYY-MM-DD): {date1_str}")

date2_str = input("Enter second date (YYYY-MM-DD): ")
print(f"Enter second date (YYYY-MM-DD): {date2_str}")

date1 = datetime.strptime(date1_str, "%Y-%m-%d")
date2 = datetime.strptime(date2_str, "%Y-%m-%d")

difference = abs((date2 - date1).days)

print(f"First date: {date1.strftime('%Y-%m-%d')}")
print(f"Second date: {date2.strftime('%Y-%m-%d')}")
print(f"Number of days between the two dates: {difference}")
```

#### Q6. Create a class `Triangle` with attributes for three angles and a method to check if their sum equals 180.

```python
class Triangle:
    def __init__(self, angle1, angle2, angle3):
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3

    def check_angles(self):
        total = self.angle1 + self.angle2 + self.angle3
        print(f"Angle 1: {self.angle1}")
        print(f"Angle 2: {self.angle2}")
        print(f"Angle 3: {self.angle3}")
        print(f"Sum of angles: {total}")

        if total == 180:
            print(f"The angles form a valid triangle.")
        else:
            print(f"The angles do not form a valid triangle.")

t1 = Triangle(60, 60, 60)
t1.check_angles()
```

#### Q7. Define a class `Song` that stores lyrics in a list and prints each line using a method `sing_me_a_song()`.

```python
class Song:
    def __init__(self, lyrics):
        self.lyrics = lyrics

    def sing_me_a_song(self):
        print(f"Printing song lyrics:")
        for line in self.lyrics:
            print(f"{line}")

happy_bday = Song([
    "Happy birthday to you",
    "I don't want to get sued",
    "So I'll stop right there"
])

happy_bday.sing_me_a_song()
```

#### Q10. Write a program using classes and objects to find the area and perimeter of a rectangle.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

length = float(input("Enter length: "))
print(f"Enter length: {length}")

width = float(input("Enter width: "))
print(f"Enter width: {width}")

rect = Rectangle(length, width)

print(f"Length: {rect.length}")
print(f"Width: {rect.width}")
print(f"Area of rectangle: {rect.area()}")
print(f"Perimeter of rectangle: {rect.perimeter()}")
```

#### Q11. Case Study: Bank Account Management (Menu Driven Program)
#### Create a class `BankAccount` with:
- account number
- name
- balance
#### Implement functions for deposit, withdraw, and display account details.

```python
class BankAccount:
    def __init__(self, account_number, name, balance):
        self.account_number = account_number
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}")
        print(f"Updated Balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient balance")
        else:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
            print(f"Updated Balance: {self.balance}")

    def display(self):
        print(f"\nAccount Details")
        print(f"Account Number: {self.account_number}")
        print(f"Name: {self.name}")
        print(f"Balance: {self.balance}")

account_number = input("Enter account number: ")
print(f"Enter account number: {account_number}")

name = input("Enter account holder name: ")
print(f"Enter account holder name: {name}")

balance = float(input("Enter initial balance: "))
print(f"Enter initial balance: {balance}")

account = BankAccount(account_number, name, balance)

while True:
    print(f"\n--- Bank Account Menu ---")
    print(f"1. Deposit")
    print(f"2. Withdraw")
    print(f"3. Display Account Details")
    print(f"4. Exit")

    choice = input("Enter your choice: ")
    print(f"Enter your choice: {choice}")

    if choice == "1":
        amount = float(input("Enter amount to deposit: "))
        print(f"Enter amount to deposit: {amount}")
        account.deposit(amount)

    elif choice == "2":
        amount = float(input("Enter amount to withdraw: "))
        print(f"Enter amount to withdraw: {amount}")
        account.withdraw(amount)

    elif choice == "3":
        account.display()

    elif choice == "4":
        print(f"Exiting program.")
        break

    else:
        print(f"Invalid choice. Please try again.")
```

#### Q1. Write a function to compute 5/0 and use try/except to catch the exception.


```python
def divide():
    try:
        result = 5/0
        print(result)
    except ZeroDivisionError:
        print("Cannot divide by zero...")

divide()
```

#### Q3. Write a program that asks the user for a file name and opens it. Handle FileNotFoundError and ValueError.

```python
try:
    filename = input("Enter file name: ")
    print(f"Enter file name: {filename}")
    if not filename:
        raise ValueError("Filename cannot be empty...")

    file = open(filename, "r")
    print(file.read())
    file.close()

except FileNotFoundError:
    print("File not found...")
except ValueError as ve:
    print(f"ValueError: {ve}")
```

#### Q4. Create a custom exception `InvalidAgeError` raised when age < 0 or > 150.


```python
class InvalidAgeError(Exception):
    pass

def check_age(age):
    if age < 0 or age > 150:
        raise InvalidAgeError("Age must be between 0 and 150...")
    else:
        print(f"Valid age: {age}")

try:
    age = int(input("Enter age: "))
    print(f"Enter age: {age}")
    check_age(age)
except InvalidAgeError as e:
    print(f"InvalidAgeError: {e}")
```

#### Q5. Write a program that performs division and handles ZeroDivisionError using try/except/else/finally.

```python
try:
    a = int(input("Enter first number: "))
    print(f"Enter first number: {a}")
    b = int(input("Enter second number: "))
    print(f"Enter second number: {b}")
    result = a / b
except ZeroDivisionError:
    print("Cannot divide by zero...")
else:
    print(f"Result is: {result}")
finally:
    print("Execution completed...")
```

#### Q7. Write a function that raises an exception and another function that catches it and raises a new exception with additional information.

```python
def first_function():
    raise ValueError("Error from first function.")

def second_function():
    try:
        first_function()
    except ValueError as e:
        raise Exception("Second function caught it and added more info.") from e

try:
    second_function()
except Exception as e:
    print(f"Caught: {e}")
```

#### Q9. Write a program that reads a file line by line and converts each line to an integer using nested try-except.

```python
try:
    filename = input("Enter file name: ")
    print(f"Enter file name: {filename}")
    file = open(filename, "r")

    for line in file:
        try:
            number = int(line.strip())
            print(number)
        except ValueError:
            print("Invalid integer in line:", line.strip())
    file.close()

except FileNotFoundError:
    print("File not found.")
```

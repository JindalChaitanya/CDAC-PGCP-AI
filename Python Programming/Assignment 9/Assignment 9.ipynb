#### Q1. Write a function to compute 5/0 and use try/except to catch the exception.


```python
def divide():
    try:
        result = 5/0
        print(result)
    except ZeroDivisionError:
        print("Cannot divide by zero...")

divide()
```

#### Q2. Write a Python program to access an array element whose index is out of bounds and handle the exception.

```python
arr = [10, 20, 30]

try:
    print(arr[5])
except IndexError:
    print("Index is out of bounds...")
```

#### Q3. Write a program that asks the user for a file name and opens it. Handle FileNotFoundError and ValueError.

```python
try:
    filename = input("Enter file name: ")
    print(f"Enter file name: {filename}")
    if not filename:
        raise ValueError("Filename cannot be empty...")

    file = open(filename, "r")
    print(file.read())
    file.close()

except FileNotFoundError:
    print("File not found...")
except ValueError as ve:
    print(f"ValueError: {ve}")
```

#### Q4. Create a custom exception `InvalidAgeError` raised when age < 0 or > 150.


```python
class InvalidAgeError(Exception):
    pass

def check_age(age):
    if age < 0 or age > 150:
        raise InvalidAgeError("Age must be between 0 and 150...")
    else:
        print(f"Valid age: {age}")

try:
    age = int(input("Enter age: "))
    print(f"Enter age: {age}")
    check_age(age)
except InvalidAgeError as e:
    print(f"InvalidAgeError: {e}")
```

#### Q5. Write a program that performs division and handles ZeroDivisionError using try/except/else/finally.

```python
try:
    a = int(input("Enter first number: "))
    print(f"Enter first number: {a}")
    b = int(input("Enter second number: "))
    print(f"Enter second number: {b}")
    result = a / b
except ZeroDivisionError:
    print("Cannot divide by zero...")
else:
    print(f"Result is: {result}")
finally:
    print("Execution completed...")
```

#### Q6. Convert a string into an integer and handle ValueError and TypeError.

```python
try:
    s = input("Enter a number: ")
    print(f"Enter a number: {s}")
    num = int(s)
    print(f"Integer value:{num}")
except ValueError:
    print("ValueError: Invalid string for integer conversion.")
except TypeError:
    print("TypeError: Wrong type given.")
```

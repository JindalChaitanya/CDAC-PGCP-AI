#### Q2: Write a Python program to get the largest number from a list.

```python
import random as r

def largest(numbers):
    largest_num = numbers[0]

    for i in numbers:
        if i > largest_num:
            largest_num = i
        else:
            continue
    
    print(f"The Lagest number from the list {numbers} = {largest_num}")

list = [r.randint(0, 1000) for _ in range(10)]
largest(list)
```

#### Q16: Write a Program to sort alphabetically the words form a string provided by the user. 
    [You can use split() method to split string into a list of words. ]


```python
string = input("Enter a string: ")

words = string.split()

for i in range(len(words)):
    for j in range(0, len(words) - i-1):
        if words[j].lower() > words[j+1].lower():
            words[j], words[j+1] = words[j+1], words[j]

print(f"Entered string: {string} \nSorted words: {words}")
```

#### Q17: write a Program to Remove Punctuations from a String provided by the user. [Hint: use punctuation attribute of string module to get all punctuations 
        (i.e. !"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~ ) ]


```python
import string

str1 = input("Enter a string with punctuations: ")

clean_str = "".join(char for char in str1 if char not in string.punctuation)

print(f"Entered String: {str1} \nOutput Cleaned String: {clean_str}")
```

#### Q18. Write a menu driven program to develop SMIS that has options to 
- accept the marks of the student in five major subjects in Class  and display the same.
- calculate the sum of the marks of all subjects. Divide the total marks by number of subjects (i.e. 5), calculate percentage = total marks/5 and display the percentage.
- Find the grade of the student as per the following criteria:

| Criteria                                 | Grade    |
|------------------------------------------|----------|
| percentage > 85                          | A        |
| percentage < 85 && percentage >= 75      | B        |
| percentage < 75 && percentage >= 50      | C        |
| percentage < 50 && percentage >= 30      | D        |
| percentage < 30                          | Reappear |


```python
marks = []
total = 0

while True:
    print("----- SMIS Menu -----")
    print("1. Accept and Display Marks")
    print("2. Calculate Total and Percentage")
    print("3. Calculate Grade")
    print("4. Exit")

    choice = int(input("Enter your choice (1-4): "))

    if choice == 1:
        marks = []
        print("Enter marks for 5 major subjects: ")
        for i in range(5):
            m = float(input(f"Subject {i+1}: "))
            marks.append(m)
            total += m
        print(f"Marks entered: {marks}")

    elif choice == 2:
        if not marks:
            print("Please enter marks first (Option 1).")
        else:
            percentage = total/ 5
            print(f"Total Marks: {total}")
            print(f"Percentage: {percentage:.2f}%")
    
    elif choice == 3:
        if not marks:
            print("Please enter marks first (Option 1).")
        else:
            if percentage >= 85:
                grade = "A"
            elif percentage >= 75:
                grade = "B"
            elif percentage >= 50:
                grade = "C"
            elif percentage >= 30:
                grade = "D"
            else:
                grade = "Reappear"
            print(f"Percentage: {percentage:.2f}% - Grade: {grade}")

    elif choice == 4:
        print("Exiting SMIS...")
        break

    else:
        print("Invalid Choice!! Please try again!")
```

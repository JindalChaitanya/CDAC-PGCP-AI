#### Q1. Write a Python program which accept the radius of a circle from the user and compute the area.

#### Q2. Temperature of a city in Fahrenheit degrees is input through the keyboard. Write a program to convert this temperature into Centigrade degrees. 

```python
f_temp = float(input("Enter the temperature in Fahrenheit: "))

c_temp = round((f_temp - 32) * (5/9), 2)

print(f"The temperature in Fahrenheit: {f_temp} is equal to {c_temp} in Celsius")
```

#### Q3. Ramesh’s basic salary is input through the keyboard. His dearness allowance is 40% of basic salary, and house rent allowance is 20% of basic salary. Write a program to calculate his gross salary. 

```python
basic_sal = int(input("Enter the basic salary: "))
dearness_allowance = basic_sal * 0.4
house_rent_allowance = basic_sal * 0.2
gross_salary = basic_sal + dearness_allowance + house_rent_allowance
print(f'''For the base salary of {basic_sal}
House Rent Allowance is {house_rent_allowance}
Dearness Allowance is {dearness_allowance}
Total Salary is {gross_salary}''')
```

#### Q8: Write a Python Program to Solve the quadratic equation ax**2 + bx + c = 0
        Coeffients a, b and c are provided by the user
        [Hint: import complex math module - import cmath]


```python
import cmath

a = float(input("Enter the coefficient of x^2 (a): "))
b = float(input("Enter the coefficient of x (b): "))
c = float(input("Enter the constant (c): "))


if a == 0:
    print("Please try again with a quadratic equation...")
else:
    discriminant = (b**2) - (4*a*c)

    root1 = ((-b) - cmath.sqrt(discriminant)) / (2*a)
    root2 = ((-b) + cmath.sqrt(discriminant)) / (2*a)

    print("\nThe solutions are:")
    print(f"Root 1 is: {root1}")
    print(f"Root 2 is: {root2}")
```

#### Q9: Write a Python Program to find the area of triangle
        Three sides of the triangle a, b and c are provided by the user


```python
import math

a = float(input("Enter the length of side 'a' of the triangle: "))
b = float(input("Enter the length of side 'b' of the triangle: "))
c = float(input("Enter the length of side 'c' of the triangle: "))

s = (a+b+c)/2

area = math.sqrt(s*(s-a)*(s-b)*(s-c))

print(f"The area of the triangle with sides {a}, {b}, {c} are: {area} unit sq.")
```

#### Q10: If a five-digit number is input through the keyboard, write a program to calculate the sum of its digits.
        (Hint: Use the modulus operator ‘%’) 

```python
num = int(input("Enter a 5-digit number: "))

def SumOfDigits(x):
    sum = 0
    while x>0:
        last_digit = x % 10
        sum += last_digit
        x = x // 10
    return sum

if num >= 10000 and num <= 99999:
    sum = SumOfDigits(num)
    print(f"The sum of digits of number {num} is: {sum}")
else:
    print("Please try again, with a 5-digit number...")
```

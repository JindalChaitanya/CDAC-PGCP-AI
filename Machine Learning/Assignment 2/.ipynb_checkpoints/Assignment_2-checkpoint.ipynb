```python
import pandas as pd

sales = pd.read_csv("sales.csv")
customer = pd.read_csv("customer.csv")
products = pd.read_csv("products.csv")
```

#### Q1. List all products sold.

```python
sales["product"]
```

#### Q2. Print all unique products.


```python
sales["product"].unique()
```

#### Q3. List the details of the product monitor.


```python
sales[sales["product"] == "Monitor"]
```

#### Q4. List the details of the product with quantity greater than 1.


```python
sales[sales["qty"] > 1]
```

#### Q5. List the details of the products in store ABC.


```python
sales[sales["store"] == "ABC"]
```

#### Q6. List the details of the products in store ABC with quantity greater than 2.


```python
sales[(sales["store"] == "ABC") & (sales["qty"] > 2)]
```

#### Q7. List the details of the product monitor and RAM.


```python
sales[sales["product"].isin(["Monitor", "RAM"])]
```

#### Q8. List the quantity sold against each product.


```python
sales_with_price = sales.merge(products, on=["p_id", "product"], how="left")
sales_with_price["total_sales"] = sales_with_price["qty"] * sales_with_price["price"]

sales_with_price.groupby("product")[["qty", "total_sales"]].sum()
```

#### Q9. List the quantity and total sales against each product.


```python
sales.groupby(["product", "store"])["qty"].sum()
```

#### Q10. List the quantity sold against each product and against each store.


```python
sales.groupby(["product", "store"])["qty"].sum()
```

#### Q11. List the quantity sold against each store with total turnover of the store.


```python
sales_with_price = sales.merge(products, on=["p_id", "product"], how="left")
sales_with_price["turnover"] = sales_with_price["qty"] * sales_with_price["price"]

sales_with_price.groupby("store")[["qty", "turnover"]].sum()
```

#### Q12. List the products which are not sold.

```python
products[~products["p_id"].isin(sales["p_id"])]
```

#### Q13. List the customers who have not purchased any product.

```python
customer[~customer["c_id"].isin(sales["c_id"])]
```

#### Q2. Fill the null value with mean value in numerical columns like Applicant Income,CoApplicant Income and LoanAmount.

```python
mean_ApplicantIncome = df["ApplicantIncome"].mean()
mean_CoapplicantIncome = df["CoapplicantIncome"].mean()
mean_LoanAmount = df["LoanAmount"].mean()
```

```python
df["ApplicantIncome"] = df["ApplicantIncome"].fillna(mean_ApplicantIncome)
df["CoapplicantIncome"] = df["CoapplicantIncome"].fillna(mean_CoapplicantIncome)
df["LoanAmount"] = df["LoanAmount"].fillna(mean_LoanAmount)
```

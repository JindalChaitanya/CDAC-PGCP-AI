```python
import numpy as np
import pandas as pd

d={
    "Player": ["M S Dhoni", "Hardik", "Rahul", "Mohit Sharma", "Andre Russel", "Jaspreet", "Virat Kohli", "Rohit Sharma"],
    "Team": ["Chennai Super Kings", "Mumbai Indians", "Kings Eleven Punjab", "Gujrat Titans", "Kolkata Knight Riders", "Mumbai Indians", "Royal Challengers Bangalore", "Mumbai Indians"],
    "Category":["Batsman", "Batsman", "Batsman", "Bowler", "Bowler", "Batsman", "Batsman", "Batsman"],
    "Bidprice":[13, 11, 10, 12, 7, 10, 17, 15],
    "Runs":[4000, 2400, 3000, 1000, 1500, 3500, 5000, 3900]
    }

df = pd.DataFrame(d)
print(df)
```

#### Q2. Display all players who are categorized as Bowlers.

```python
df[df['Category'] == 'Bowler'][['Player', 'Category']]
```

#### Q3. Display all players who belong to the Mumbai Indians team.


```python
df[df["Team"] == 'Mumbai Indians'][['Player', 'Team']]
```

#### Q4. Display all Mumbai Indians players whose total runs are greater than 3000.


```python
df[(df["Team"] == 'Mumbai Indians') & (df["Runs"] >= 3000)][['Player', 'Team', 'Runs']]
```

#### Q5. Display the complete details of players who play for either RCB or Chennai Super King.


```python
df[(df['Team'] == 'Chennai Super Kings') | (df['Team'] == 'Royal Challengers Bangalore')]
```
